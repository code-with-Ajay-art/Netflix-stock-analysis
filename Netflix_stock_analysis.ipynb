# Netflix stock price analysis using Python
##  Objective
To analyze Netflix’s historical stock performance using Python and visualize trends, trading volume, moving averages, and daily returns.

## Step 1 : Import required libraries
## Step 2 : Load and preview dataset
## Step 3 : convert date and set as index


```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\HP\Downloads\archive (1)\Netflix_stock_data.csv")
print(df.columns)
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date',inplace=True) # set 'date' as index to make time series analysis easy


```

## Step 4 : View basic information and summary statistics

```python
df.head(5) #  showing the first five rows of data
df.info() #show data type , no. of null values and memory ussage
df.describe() # give summary of statistics
```

# step 4.1 Filtering the data

```python
df_2023 =df.loc['2023-01-01':'2023-12-31'] # filter data for 2023 only
df_2023.tail(5) # show last five rows
```

##  Step 5: Visualize Closing Stock Price

```python
 

plt.plot(df['Close'],color='black',linestyle='-',label='close price',marker='')
plt.title('Netflix closing stock price over time')
plt.xlabel('Date')
plt.ylabel('Close')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

##  Step 6: Add and Plot 20 & 50 Day Moving Averages


```python
df['MA20']= df['Close'].rolling(20).mean()
df['MA50']= df ['Close'].rolling(50).mean()

plt.plot(df['Close'],alpha=0.5)

plt.plot(df['MA20'],color='orange',label='MA20')
plt.plot(df['MA50'],color='green',label='MA50')
plt.title('Netflix close price with 20 and 50 day moving average')
plt.xlabel('Date')
plt.ylabel('Price($)')
plt.legend()

plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

##  Step 7: Trading Volume Over Time


```python
plt.plot(df['Volume'],color='purple')
plt.title('Netflix trading volume over time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.grid()
plt.tight_layout()
plt.show()
```

## Step 8: Daily Return (%) Analysis


```python
df['Daily_returns_(%)']= df['Close'].pct_change() *100
plt.figure(figsize= (16,9))
plt.plot(df['Daily_returns_(%)'],color='red')
plt.title('Netflix daily return percentage')
plt.xlabel('Date')
plt.ylabel('Daily_returns_(%)')
plt.tight_layout()
plt.show()
```

##  Final Insights
- Netflix showed strong price volatility during market crashes and recovery periods.
- Moving averages smoothed noise and highlighted trend direction.
- Volume spikes aligned with key market actions.
- Daily return % helped track risk and short-term change.
